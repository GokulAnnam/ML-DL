```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 'year' and 'per capita income (US$)'
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])

# Predict the value for 2005
new_year = 2005
predicted_value = reg.predict([[new_year]])

# Create a scatter plot for the original data
plt.scatter(df['year'], df['per capita income (US$)'], color='blue', label='Original Data')

# Add the predicted value for 2005 as a red point
plt.scatter(new_year, predicted_value, color='red', label=f'Prediction for {new_year}')

# Adding labels and title
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Per Capita Income with Prediction for 2005")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
years = [2005, 2001, 1997, 2002,1999]
years = np.array(years).reshape(5,1)
predicted_value = reg.predict(years)
plt.scatter(df['year'],df['per capita income (US$)'],color = 'cyan',label = f'original')
plt.scatter([year[0] for year in years], predicted_value, color='red', label=f'Prediction for {new_year}')
plt.legend()
```
